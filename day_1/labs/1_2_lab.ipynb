## Matplotlib practice

```python
# Create a Dataframe for the function f(x) = x^2 + 2x + 1 for x = 1 to 10 (you can choose step size, e.g., 1)
x = np.arange(0, 10., 0.1)
y = 2 ** x + 2 * x + 1
df_math = pd.DataFrame()
df_math['x'] = x
df_math['y'] = y

# Plot the DataFrame representing y = 3x + 2

plt.figure(figsize=(10, 5))
plt.plot(df_math['x'], df_math['y'], label='f(x) = x^2 + 2x + 1', color='blue')
plt.title('Plot of f(x) = x^2 + 2x + 1')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

```python
# Plot the y column against x (from the dataframe created above) using matplotlib
x = [1, 2, 3, 4, 5]
y = np.sin(x) + np.cos(x)
xy = pd.DataFrame()
xy['x'] = x
xy['y'] = y

# Plot the DataFrame representing y = 3x + 2

plt.figure(figsize=(10, 5))
plt.plot(xy['x'], xy['y'], label='Plot, color='blue')
plt.title('Plot!')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```
