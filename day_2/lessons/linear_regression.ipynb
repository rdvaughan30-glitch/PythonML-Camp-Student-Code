```python
# Import necessary libraries
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read synthetic data from CSV file, get X and y
data = pd.read_csv('data/regression_data.csv')
X = data[['x']].to_numpy()
y = data['y'].to_numpy()
```

```python
# Check the shapes of X and y, notice that the shape of X is (# of data points, # of features) and the shape of y is (# of data points,)
X.shape, y.shape
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(X, y, color='blue', label='Data points')
plt.xlabel(('x'))
plt.ylabel(('y'))
plt.title('Synthetic Data')
plt.legend()
plt.show()
```

```python
# Create and train the linear regression model, get the estimated parameters
model = LinearRegression()
model.fit(X, y)
a_est = model.coef_[0]
b_est = model.intercept_

print(f'Estimated intercept: {a_est:.2f}')
print(f'Estimated slope: {b_est:.2f}')
```

```python
# generate predictions
y_pred = model.predict(X)

plt.figure(figsize=(5,5))
plt.scatter(X, y, label='data points')
plt.plot(X, y_pred, label='fitted line', linewidth=2, color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Fit')
plt.legend()
plt.show()
# plot data and fitted line
```

```python
# Calculate and print the R-squared value
r_squared = model.score(X, y)
print(f'R-squared: {r_squared:.2f}')
```
